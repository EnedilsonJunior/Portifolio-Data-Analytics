# Somente uma pasta - 2012


```python
import pandas as pd
import glob
import os

# Defina o diretório que contém as planilhas
def list_sheets():
    directory = r'C:\Users\enedilson\Desktop\Python\Portifólio\SCR\planilha_2012'
    arquivos = os.listdir(directory)
    imagens_processadas = []

    for arquivo in arquivos:
        if arquivo.lower().endswith('.csv'):
            caminho_completo = os.path.join(directory, arquivo)
            imagens_processadas.append(caminho_completo)

    return imagens_processadas

arquivos = list_sheets()
# Crie uma lista para armazenar as planilhas
planilhas = []

print(arquivos)

# Leia cada arquivo e adicione à lista
i = 0

# planilha = pd.read_csv("planilha_2012/planilha_201207.csv" , sep=';')
# display(planilha)

for arquivo in arquivos:
    planilha = pd.read_csv(arquivo, sep=';')
    #display(planilha)
    print("Planilha "+str(i))
    #display(planilha)
    i += 1
    planilhas.append(planilha)

# Agrupe as planilhas em uma única planilha
planilha_agrupada = pd.concat(planilhas, ignore_index=True)

display(planilha_agrupada)

# # Salve a planilha agrupada em um novo arquivo
planilha_agrupada.to_csv('planilha_agrupada.csv', index=False)# Somente uma pasta - 2012


```
