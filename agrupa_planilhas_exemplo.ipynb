```python
import os
import pandas as pd
```

```python
def list_sheets():
    directory = r"Z:\ssti\danpp\sbdata\00-Serviços\41985 - BI atendimentos suporte 1Doc\Tickects"
    arquivos = os.listdir(directory)
    imagens_processadas = [arquivo for arquivo in arquivos if arquivo.lower().endswith('.xlsx')]
    return imagens_processadas
```

```python
all_sheets = list_sheets()
atendimento = pd.DataFrame()
directory = r"Z:\ssti\danpp\sbdata\00-Serviços\41985 - BI atendimentos suporte 1Doc\Tickects\Tickets Juiz de Fora_MG (11).xlsx"
all = pd.ExcelFile(r"Z:\ssti\danpp\sbdata\00-Serviços\41985 - BI atendimentos suporte 1Doc\Tickects\Tickets Juiz de Fora_MG (11).xlsx")

sheet=all.sheet_names

print(sheet)


```

```python
for sht in sheet:   
    sheet_data = pd.read_excel(directory, sheet_name=sht)
    atendimento = pd.concat([atendimento, sheet_data], ignore_index=True)
```
